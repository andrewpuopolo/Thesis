```python
def loglik(k, initHF):
    savedELOs=[]
    initdictionary={}
    h=initHF
    inithome=[1750+initHF,1500+initHF,1250+initHF,1000+initHF]
    initaway=[1750-initHF,1500-initHF,1250-initHF,1000-initHF]
    initialrows=[]

    #This needs to be rewritten so we don't have to run it over an dover again
    for i in range(len(fulldf['home'].unique())):
        Team=fulldf['home'].unique()[i]
        Teamdf=fulldf[fulldf['home']==Team].reset_index()
        Tier=int(Teamdf['tier'][0])
        initialrow=[Team, Tier]
        initialrows.append(initialrow)
    #Needs to be rewritten to optimize using BuildEloDict Function
    for a in range(len(initialrows)):
        initdictionary[initialrows[a][0]]=[inithome[initialrows[a][1]-1], initaway[initialrows[a][1]-1]]
    #print(initdictionary)
    #avgE=0
    #HFA=initHF
    logliks=0.0
    burn_in=40
    Elodictionary=initdictionary.copy()
    #startyear=fulldf['Season'].values[0]
    #HFAyear=fulldf['Season'].values[0]
    #HFcounter=0.
    #Gamecounter=0.
    #Rs=0
    #logcount=0
    #HFATime=[]
    for ind, row in fulldf.iterrows():
        Home=row['home']
        Away=row['visitor']
        HomeGoals=row['hgoal']
        AwayGoals=row['vgoal']
        year=row['Season']
        HomeELO=Elodictionary[Home][0]
        AwayELO=Elodictionary[Away][1]
        Elodif=float((HomeELO-AwayELO)/400.0)
        #avgE+=Elodif
        E=1/(10**(-1*(Elodif))+1)
        #Write Lambda Functions for this to calculate Rvalue and Loglik, gets rid of if statements
        Resfun = lambda x,y: (x > y)*1 + (x == y)*(.5)+(x<y)*0
        R=Resfun(HomeGoals, AwayGoals)
        mloglikfun = lambda x,y: np.log(np.sqrt(y*(1-y))) if x==.5 else np.log((-1*x)+1+(2*x-1)*y)
        MatchLogLik= mloglikfun(R, E)
        #if HomeGoals>AwayGoals:
         #   R=1
          #  MatchLogLik=np.log(E)
       # elif HomeGoals<AwayGoals:
        #    R=0
         #   MatchLogLik=np.log(1-E)
       # else:
        #    R=.5
         #   MatchLogLik=np.log(np.sqrt(E*(1-E)))
        PointExchange=(R-E)*k
        Elodictionary[Home][0]=HomeELO+PointExchange
        Elodictionary[Away][1]=AwayELO-PointExchange
        
        if(year>1958+burn_in):
            logliks+=MatchLogLik
        


    print([k, initHF, np.sum(logliks), np.mean(logliks)])
    #print(Elodictionary)
    
    return -1*np.sum(logliks)
```

```python
from scipy.optimize import minimize
#mins=[]
for i in range(10):
    initk=np.random.uniform(0,100)
    inithf=np.random.uniform(0,200)
    x0=[initk,inithf]
    res=minimize(lambda x: loglik(*x),x0, method='SLSQP', bounds=[[0,300], [0,2000]],
                 options={'disp': True})
    row=[i, initk, inithf,  res['x'][0], res['x'][1], res['fun']]
    print(row)
    mins.append(row)
```
